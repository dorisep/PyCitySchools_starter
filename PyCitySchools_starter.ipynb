```python
# Dependencies and Setup
import pandas as pd
import numpy as np

# File to Load (Remember to Change These)
school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

# Read School and Student Data File and store into Pandas Data Frames
school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

# Combine the data into a single dataset
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
```

```python
# Calculate the number of schools
school_count = len(school_data_complete["school_name"].unique())
# Calculate the number of students
student_count = (school_data_complete["Student ID"].nunique())
# Calculate the total Budget
budg_df = school_data_complete[["school_name","budget"]]
new_budg_df = budg_df.drop_duplicates()
total_budget = new_budg_df["budget"].sum()                     
# avg math score
avg_math =int(round(school_data_complete["math_score"].mean()))
# avg reading score
avg_read = int(round(school_data_complete["reading_score"].mean()))

# percent math pass 
bins = [0, 69, 100]
group_names = ["fail", "pass"]
math_pass = pd.cut(school_data_complete["math_score"], bins, labels=group_names)
math_pass_perc = int(round(math_pass.value_counts()["pass"]/student_count*100))

# percent read pass
bins = [0, 69, 100]
group_names = ["fail", "pass"]
read_pass = pd.cut(school_data_complete["reading_score"], bins, labels=group_names)
read_pass_perc = int(round(read_pass.value_counts()["pass"]/student_count*100))

# overall passing rate
overall_pass = int(round(math_pass_perc + read_pass_perc) / 2)
read_pass_perc
```

```python
dist_summ_dic = ({"Total Schools": school_count, "Total Students": student_count,
                  "Total Budget": total_budget, "Average Math Score":  avg_math, 
                  "Average Reading Score": avg_read, "Math Pass Rate": math_pass_perc, 
                  "Reading Pass Rate": read_pass_perc, "Passing Rate": overall_pass})
district_summary_df = pd.DataFrame([dist_summ_dic])
district_summary_df = district_summary_df [["Total Schools","Total Students", "Total Budget", "Average Math Score", 
                                            "Average Reading Score", "Math Pass Rate", "Reading Pass Rate", 
                                            "Passing Rate"]]
district_summary_df
```

```python
school_data_complete.head()

school_df = school_data.set_index('school_name')

df_district = school_df['type']
df_district


total_students = school_data_complete.groupby('school_name').count()
total_students = total_students['student_name']
total_students


total_school_budget = school_data_complete.groupby('school_name').mean()['budget']
total_school_budget



per_student_budget = total_school_budget / total_students
per_student_budget

stu_avg_math = school_data_complete.groupby('school_name').mean()['math_score']
stu_avg_math  

stu_avg_read = school_data_complete.groupby('school_name').mean()['reading_score']
stu_avg_read

stu_pass_math = school_data_complete[ school_data_complete['math_score'] >= 70]

percent_pass_math = (stu_pass_math.groupby('school_name').count()['math_score'] / school_data_complete.groupby('school_name').count()['math_score']) * 100 

stu_pass_read = school_data_complete[ school_data_complete['reading_score'] >= 70]
percent_pass_read = (stu_pass_read.groupby('school_name').count()['reading_score'] / school_data_complete.groupby('school_name').count()['reading_score']) * 100 



overall_pass = (percent_pass_math + percent_pass_read) / 2

school_summary_df = pd.DataFrame({'school_type': df_district,
                            'student_pop': total_students,
                            'school_budget': total_school_budget,
                              'per_student_budget' : per_student_budget,
                                  'student_math_avg' : stu_avg_math,
                                'student_reading_avg' : stu_avg_read,
                                'percent_pass_math' : percent_pass_math,
                                'percent_pass_read' : percent_pass_read,
                                  'overall_pass' : overall_pass

                                 })
school_summary_df
```
